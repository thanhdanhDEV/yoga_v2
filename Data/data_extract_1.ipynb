```python
path_x = 'landmarks.csv'
path_y = 'labels.csv'
df_x = pd.read_csv(path_x)
df_y = pd.read_csv(path_y)


df_x.drop(columns=["pose_id"], inplace=True)
df_y.drop(columns=["pose_id"], inplace=True)
```

```python
data_pushups_up = df_x.loc[df_y.pose == 'pushups_up']
data_pushups_down = df_x.loc[df_y.pose == 'pushups_down']

label_pushups_up = df_y.loc[df_y.pose == 'pushups_up']
label_pushups_down = df_y.loc[df_y.pose == 'pushups_down']
```

```python

X = pd.concat([data_pushups_up, data_pushups_down], ignore_index=True)
y = pd.concat([label_pushups_up, label_pushups_down], ignore_index=True)

y.replace("pushups_up", "1", inplace= True)
y.replace("pushups_down", "0", inplace= True)
```

```python
X_input = X.to_numpy(dtype= np.float32)
y_input = y.to_numpy(dtype= np.uint8)
```

```python
#### X + y -> Data frames
df = pd.concat([X, y], axis= 1)
```

```python
### Save to csv
import os  
os.makedirs('Data_CSV', exist_ok=True)  
df.to_csv('Data_CSV/landmarks.csv', index=False)  
```
